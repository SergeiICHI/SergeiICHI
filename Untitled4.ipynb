```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
sns.scatterplot(data=df, x = 'longitude', y = 'latitude') #по оси  x и y задаем амплитуды
```

```python
sns.scatterplot(data=df, x = 'longitude', y = 'latitude', hue = 'median_house_value') # по оси x y задаем амплитуды + добавляем стоимость в цвете
```

```python
cols = ['population', 'median_income', 'housing_median_age', 'median_house_value']
g = sns.PairGrid(df[cols])
g.map(sns.scatterplot) #PairGrid транслирует все графики
```

```python
sns.relplot(data = df, x='latitude', y='median_house_value', kind='line') #график relplot
```

```python
sns.histplot(data = df, x = 'population') # график hisplot
```

```python
new = df[df['population'] < 6000]  # создаем новое значение у population чтобы посмотреть углубленно на график
```

```python
sns.histplot(data = new, x = 'population') # приминяем новое значение
```

```python

df.loc[df['housing_median_age'] <= 20, 'age_group'] = 'Молодые'
df.loc[(df['housing_median_age'] > 20) & (df['housing_median_age'] <= 50), 'age_group'] = 'Cp.возраст'
df.loc[df['housing_median_age'] > 50, 'age_group'] = 'Пожилые'

df.groupby('age_group')['median_income'].mean().plot(kind='bar')

#разделяем на группы

```
